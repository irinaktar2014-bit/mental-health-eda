## Mental Health in Tech - Exploratory Data Analysis (EDA)

## Project Objective
The goal of this project is to explore mental health trends among employees in the tech industry. 
Using survey data, we analyze demographic factors, workplace support, and treatment-seeking behavior to understand how mental health impacts employees.

## Dataset
Source: Kaggle - Mental Health in Tech Survey
Rows: 1,259
Columns: 27

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("mental_health_eda.ipynb.csv.csv")
df.head()
```

```python
df["Age"].describe()
df = df[(df["Age"] > 10) & (df["Age"] < 100)]
df["work_interfere"].fillna("Unknown", inplace=True)
df.isnull().sum()
```

```python
df["Gender"].value_counts().head(10)
df["Gender"].value_counts().head(10).plot(kind="bar")
plt.title("Gender Distribution")
plt.show()
```

```python
plt.hist(df["Age"], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
df["treatment"].value_counts()
df["treatment"].value_counts().plot(kind='bar')
plt.title("Mental Health Treatment")
plt.show()
```

```python
df["work_interfere"].value_counts()
pd.crosstab(df["family_history"], df["treatment"])
pd.crosstab(df["family_history"], df["treatment"]).plot(kind='bar')
plt.title("Family History vs Treatment")
plt.show()
```

## Key Insights
- Most respondents fall within the age range of 25-40.
- A large number of employees have sought mental health treatment.
- Employees with a family history of mental illness are more likely to seek treatment.
- Mental health issues often interfere with work performance.
- Many companies do not provide clear mental health benefits.
    

## Conclusion
This exploratory data analysis shows that mental health is a significant concern among employees. The findings suggest that organizations should improve mental health benefits, reduce stigma, and create supportive workplace environments.
Better awareness and policies can lead to healthier and more productive employees.
